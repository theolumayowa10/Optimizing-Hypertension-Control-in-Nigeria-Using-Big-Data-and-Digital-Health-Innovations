# Hypertension Prevalence Data Analysis Project (Focus: Nigeria)

##  Introduction

Hypertension, commonly referred to as high blood pressure, is a major risk factor for cardiovascular diseases and one of the leading causes of premature death globally. It affects over a billion people worldwide, with its prevalence rising particularly in low- and middle-income countries, where health systems often struggle to manage non-communicable diseases effectively.

Nigeria, as Africa’s most populous country, faces a significant burden of hypertension-related health challenges. Despite ongoing public health efforts, gaps remain in understanding the detailed prevalence trends across demographic groups and over time. Moreover, Nigeria's position relative to other countries and regions in terms of hypertension prevalence is not always clearly understood.

For this project, the dataset was **scraped directly from the official website of the World Health Organization (WHO)** using **Chat4Data**, an automated data scraping tool designed for structured data extraction from online sources. This approach ensured the use of accurate, globally recognized data for the analysis.

This project aims to address these knowledge gaps by analyzing a comprehensive global hypertension dataset, focusing specifically on Nigeria, to uncover temporal trends, demographic disparities, and regional comparisons. The insights derived can support data-driven policy-making and public health strategies targeted at reducing the national burden of hypertension.

---

##  Problem Statement

**Understanding Hypertension Prevalence in Nigeria: A Demographic, Temporal, and Comparative Analysis**

Hypertension continues to pose a significant public health challenge worldwide, contributing substantially to cardiovascular morbidity and mortality. Despite global health initiatives, many countries, including Nigeria, struggle with high hypertension prevalence rates due to socioeconomic, lifestyle, and healthcare access factors.

While general statistics on hypertension in Nigeria exist, a critical gap remains in understanding how prevalence varies across demographic subgroups (especially between sexes), how it evolves over time, and how Nigeria compares to global and regional benchmarks. This lack of granular, data-driven insights limits the effectiveness of targeted interventions and policy decisions.

This project seeks to address these gaps by systematically analyzing Nigeria’s hypertension prevalence within a broader global context, providing a clearer picture of demographic disparities, temporal patterns, and Nigeria’s standing relative to other countries.

# Insights

### 1. Global average hypertension rate over time

```python
time_trend = df.groupby('DIM_TIME')['RATE_PER_100_N'].mean()

# Plot
time_trend.plot(title='Global Average Hypertension Rate Over Time', figsize=(10, 5))
plt.xlabel('Year')
plt.ylabel('Prevalence Rate per 100')
plt.grid(True)
plt.show()
```

### 2. Sex Based Comparison

```python
sex_comparison = df.groupby('DIM_SEX')['RATE_PER_100_N'].mean()

sex_comparison.plot(kind='bar', color=['salmon', 'skyblue', 'lightgreen'])
plt.title('Average Hypertension Rate by Sex')
plt.ylabel('Prevalence Rate per 100')
plt.grid(axis='y')
plt.show()
```

### 3. Top 20 countries by prevalemce

```python
country_avg = df.groupby('GEO_NAME_SHORT')['RATE_PER_100_N'].mean().sort_values(ascending=False)

country_avg.head(20).plot(kind='barh', figsize=(8, 10))
plt.xlabel('Average Hypertension Prevalence')
plt.title('Top 20 Countries by Hypertension Prevalence')
plt.gca().invert_yaxis()
plt.show()
```

### 3. Confidence Interval Width Analysis

```python
df['CI_Width'] = df['RATE_PER_100_NU'] - df['RATE_PER_100_NL']

df['CI_Width'].hist(bins=30)
plt.title('Distribution of Confidence Interval Width')
plt.xlabel('Width of Confidence Interval')
plt.ylabel('Frequency')
plt.show()
```

### 4. Country-Specific Trend (Nigeria)

```python
nigeria_data = df[df['GEO_NAME_SHORT'] == 'Nigeria']

nigeria_data.groupby('DIM_TIME')['RATE_PER_100_N'].mean().plot(marker='o')
plt.title('Hypertension Prevalence in Nigeria Over Time')
plt.xlabel('Year')
plt.ylabel('Prevalence Rate per 100')
plt.grid(True)
plt.show()
```

### 5. Male vs. Female Hypertension Prevalence in Nigeria

```python
nigeria_sex_comparison = nigeria_data.groupby('DIM_SEX')['RATE_PER_100_N'].mean()

nigeria_sex_comparison.plot(kind='bar', color=['salmon', 'skyblue', 'lightgreen'])
plt.title('Average Hypertension Rate in Nigeria by Sex')
plt.ylabel('Prevalence Rate per 100')
plt.grid(axis='y')
plt.show()
```

### 6. Confidence Interval Width Over Time (Data Certainty Check)

```python
nigeria_data = df[df['GEO_NAME_SHORT'] == 'Nigeria'].copy()
nigeria_data['CI_Width'] = nigeria_data['RATE_PER_100_NU'] - nigeria_data['RATE_PER_100_NL']
df['CI_Width'] = df['RATE_PER_100_NU'] - df['RATE_PER_100_NL']

df['CI_Width'].hist(bins=30)
plt.title('Distribution of Confidence Interval Width')
plt.xlabel('Width of Confidence Interval')
plt.ylabel('Frequency')
plt.show()
```

### 7. Highest Recorded Hypertension Rate in Nigeria

```python
highest_rate_row = nigeria_data.loc[nigeria_data['RATE_PER_100_N'].idxmax()]
highest_rate_row[['DIM_TIME', 'DIM_SEX', 'RATE_PER_100_N']]
```

### 8. Trend of Minimum, Average, and Maximum Prevalence Rates Over Time

```python
nigeria_stats = nigeria_data.groupby('DIM_TIME')['RATE_PER_100_N'].agg(['min', 'mean', 'max'])

nigeria_stats.plot(marker='o')
plt.title('Min, Average, and Max Hypertension Rates in Nigeria Over Time')
plt.xlabel('Year')
plt.ylabel('Prevalence Rate per 100')
plt.grid(True)
plt.legend(['Min', 'Mean', 'Max'])
plt.show()
```

### 9. Comparing Nigeria vs. Global Average Over Time

```python
global_trend = df.groupby('DIM_TIME')['RATE_PER_100_N'].mean()
nigeria_trend = nigeria_data.groupby('DIM_TIME')['RATE_PER_100_N'].mean()

plt.plot(global_trend.index, global_trend.values, label='Global Average', linestyle='--')
plt.plot(nigeria_trend.index, nigeria_trend.values, label='Nigeria', marker='o')

plt.title('Nigeria vs. Global Average Hypertension Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Prevalence Rate per 100')
plt.legend()
plt.grid(True)
plt.show()
```

### 10. Nigeria vs. Africa Average

```python
african_countries = df[df['DIM_GEO_CODE_M49'].between(2, 899)]
africa_trend = african_countries.groupby('DIM_TIME')['RATE_PER_100_N'].mean()

plt.plot(africa_trend.index, africa_trend.values, label='Africa Average', linestyle='--')
plt.plot(nigeria_trend.index, nigeria_trend.values, label='Nigeria', marker='o')

plt.title('Nigeria vs. Africa Average Hypertension Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Prevalence Rate per 100')
plt.legend()
plt.grid(True)
plt.show()
```

### 11. Nigeria vs. Top 5 Countries Globally (Latest Year)

```python
# Define latest year
latest_year = df['DIM_TIME'].max()

# Filter for latest year globally
global_latest = df[df['DIM_TIME'] == latest_year]

# Get top 5 countries globally
top5_countries = global_latest.groupby('GEO_NAME_SHORT')['RATE_PER_100_N'].mean().sort_values(ascending=False).head(5)

# Append Nigeria's average
nigeria_latest = nigeria_data[nigeria_data['DIM_TIME'] == latest_year]['RATE_PER_100_N'].mean()
top5_countries['Nigeria'] = nigeria_latest

# lot comparison
top5_countries.sort_values(ascending=True).plot(kind='barh', color='cornflowerblue')
plt.xlabel('Hypertension Rate per 100')
plt.title(f'Nigeria vs. Top 5 Countries in Hypertension Prevalence ({latest_year})')
plt.show()
```
